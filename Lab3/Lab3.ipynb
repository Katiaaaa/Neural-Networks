### sklearn.datasets.make_classification
Створіть випадкову задачу класифікації n-класу . Це спочатку створює кластери точок, нормально розподілених (std=1) навколо вершин n_інформативно-вимірного гіперкуба зі сторонами довжиною 2*class_sep і призначає рівну кількість кластерів кожному класу.
Тут ми використовуємо make_classification для створення даних класифікації. У нас є збережені функції та цілі.
* n_samples: це означає кількість зразків (рядків), які ми хочемо в нашому наборі даних. За замовчуванням встановлено значення 100
* n_features: це означає кількість функцій (стовпців), які ми хочемо в нашому наборі даних. За замовчуванням встановлено значення 20
* n_informative: Використовується для встановлення номера інформаційного класу. За замовчуванням встановлено значення 2
* n_redundant: використовується для встановлення кількості зайвих функцій. Ознаки, які можуть бути сформовані як випадкові лінійні комбінації інформаційних ознак. За замовчуванням встановлено значення 2
* n_classes : Це означає кількість класів у цільовому наборі даних.


```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import RidgeClassifier
from sklearn.metrics import balanced_accuracy_score, f1_score, log_loss, precision_score

#generation data
X, y_true = make_classification(n_samples=100, n_features=20, n_informative=2, n_redundant=2)
#modeling data
model = RidgeClassifier().fit(X, y_true)
y_pred = model.predict(X)
print(model.score(X, y_true))

# метрики якості
# обчислення збалансованої точності
print(balanced_accuracy_score(y_true, y_pred))
# обчислення збалансованого показника F
print(f1_score(y_true, y_pred, average='binary'))
# обчислення втрат логістики
print(log_loss(y_true, y_pred))
# обчислення точності
print(precision_score(y_true, y_pred, average='binary'))
```
